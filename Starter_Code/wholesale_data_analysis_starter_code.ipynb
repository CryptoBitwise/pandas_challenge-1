## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['line_sumtotal'] = df['unit_price'] * df['qty']
df.head()
#df[['category',	'subcategory',	'unit_price',	'unit_cost', 'unit_weight',	'qty', 'line_number','line_sumtotal']]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']


def calculate_shipping_price(weight):
    if weight > 50:
        return weight * 7
    else:
        return weight * 10
    


df['Shipping Price'] = df['total_weight'].apply(calculate_shipping_price)

df[['unit_weight', 'qty', 'total_weight', 'unit_price', 'Shipping Price']].head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
sales_tax_rate = 1.0925

df['total_price'] = round( (df['line_sumtotal'] + df['Shipping Price']) * sales_tax_rate, 2)
df.head()
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit.
summary_df = df.groupby('client_id').agg({
    'qty': 'sum',
    'Shipping Price': 'sum',
    'total_price': 'sum',
    'profit': 'sum'
}).reset_index()

# Rename the columns for clarity
summary_df = summary_df.rename(columns={
    'qty': 'total_units_purchased',
    'Shipping Price': 'total_shipping_price',
    'total_price': 'total_revenue',
    'profit': 'total_profit'
})

# Sort the data by total revenue in descending order and select the top 5 clients
summary_df = summary_df.sort_values(by='total_revenue', ascending=False).head(5)

print(summary_df)
```
